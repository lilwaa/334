```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
# Import dataset
df = pd.read_csv("raw-data/mentalhealth.csv", index_col=False)
# original size: 292363 x 17
```

```python
# Drop rows containing NAs
df.dropna(how='any', inplace=True)
# new size: 287162 x 17

# Subset only countries containing United States
df=df[df['Country'] == 'United States']
# new size: 168056 x 17

# Drop Unnecessary columns: Timestamp, mental health interview, country
df = df.drop('Timestamp', axis=1)
df = df.drop("mental_health_interview", axis = 1)
df = df.drop("Country", axis =1 )

df
```

```python
# Check ratios of each class
df.describe()
```

```python
# Convert each column to numeric
# Boolean columns are converted 0/1 (i.e. Gender, self_employed, family_history, treatment, Growing_stress)
df['Gender'] = np.where(df['Gender'] == 'Female', 1, 0)
df['self_employed'] = np.where(df['self_employed'] == 'Yes', 1, 0)
df['family_history'] = np.where(df['family_history'] == 'Yes', 1, 0)
df['treatment'] = np.where(df['treatment'] == 'Yes', 1, 0)
df['Growing_stress'] = np.where(df['Growing_Stress'] == 'Yes', 1, 0)
df['Coping_struggles'] = np.where(df['Coping_Struggles'] == 'Yes', 1, 0)

# One hot encode the other columns
def onehot(inputdf, colname):
    hot_encoded_data = pd.get_dummies(inputdf, columns = [colname])
    return hot_encoded_data

col_names = ['Occupation', 'Days_Indoors', 'Growing_Stress', 'Changes_Habits', 'Mental_Health_History', 'Mood_Swings', 'Coping_Struggles', 'Work_Interest', 'Social_Weakness', 'care_options']
for col in col_names: 
    df = onehot(df, col)

# 168056 rows x 39 col
```

```python
# save with date (don't accidentally rewrite files!!!)
df.to_csv('clean-data/x.csv') 
```
